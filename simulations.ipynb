```julia
function randomwalk(n,sigma)
    x = [0.0]
    nd = Normal(0,sigma)
    for i in 1:n
        push!(x,x[end] + rand(nd))
    end
    return x
end
```

```julia
n = 1000
x_msd = zeros(1001)
for i in 1:n
    x_msd = x_msd .+ randomwalk(1000,1).^2
end
x_msd = x_msd ./ n
```

```julia
plot(x_msd)
```

```julia
function ou(n,sigma)
    x = [0.0]
    nd = Normal(0,sigma)
    for i in 1:n
        push!(x,0.9*x[end] + rand(nd))
    end
    return x
end
```

```julia
x = ou(1000,1)
```

```julia
plot(x)
```

```julia
histogram(x, normed=true)
xrange = -6:0.1:10
plot!(xrange,pdf.(Normal(mean(x),std(x)),xrange))
```

```julia
function dw(n,sigma,a,b)
    x = [0.0]
    nd = Normal(0,sigma)
    for i in 1:n
        push!(x, x[end] +b*x[end]-a*x[end]^3 + rand(nd))
    end
    return x
end
```

```julia
x = dw(10000,1,0.01,0.25)
plot(x)
```
