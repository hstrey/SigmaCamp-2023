```julia
function random_walk(n, sigma)
    x = [0.0]
    nd = Normal(0, sigma)
    for i in 1:n-1
        push!(x, x[end] + rand(nd))
    end
    return x
end
```

```julia
random_walk(1000,1)
```

```julia
nm = 1000000
n = 1000
x_mean = zeros(n)
for i in 1:nm
    x_mean = x_mean .+ random_walk(n,1).^2
end
x_mean = x_mean /nm
```

```julia
plot(x_mean)
```

```julia
function ou(n, sigma)
    x = [0.0]
    nd = Normal(0, sigma)
    for i in 1:n-1
        push!(x, 0.9*x[end] + rand(nd))
    end
    return x
end
```

```julia
x = ou(1000,0.5)
```

```julia
plot(x .+ 5)
```

```julia
function ou(n, sigma, k)
    x = [0.0]
    nd = Normal(0, sigma)
    for i in 1:n-1
        push!(x, (1-k)*x[end] + rand(nd))
    end
    return x
end
```

```julia
xou = ou(1000,1,0.1)
```

```julia
plot(xou)
```

```julia
sigma_range = 0.1:0.1:5
var_list = []
for sigma in sigma_range
    xou = ou(1000,sigma,0.1)
    push!(var_list, std(xou)^2)
end
```

```julia
plot(sigma_range, sqrt.(var_list))
```

```julia
@model function singlenorm(data)
    μ ~ Uniform(-10,10)
    σ ~ Uniform(0.1,5)
#    for i in 1:length(data)
#        data[i] ~ Normal(μ,σ)
#    end
# we can do the same using a multivariate Normal
    data ~ MvNormal(ones(length(data))*μ,σ)
end
```

```julia
chain = Turing.sample(mymodel,NUTS(0.65),1000)
```

```julia
@model function gaussian_model(sample)
    μ ~ Uniform(-10,10)
    σ ~ Uniform(0.001,5)
    for i in 1:length(sample)
        sample[i] ~ Normal(μ,σ)
    end
end
```

```julia
data = rand(nd,100)
```

```julia
mymodel = gaussian_model(data)
```

```julia
chain = Turing.sample(mymodel, NUTS(0.65), 1000)
```

```julia
plot(chain[:μ])
```

```julia
histogram(chain[:σ], bins=50)
```

```julia
@model function compare_dist(sample1,sample2)
    deltamu ~ Uniform(-1,1)
    mu ~ Uniform(-10,10)
    sigma1 ~ Uniform(0.001,5)
    sigma2 ~ Uniform(0.001,5)
    for i in 1:length(sample1)
        sample1[i] ~ Normal(mu+deltamu/2,sigma1)
    end
    for i in 1:length(sample2)
        sample2[i] ~ Normal(mu-deltamu/2,sigma2)
    end
end
```

```julia
sample1 = rand(Normal(0,1),20)
```

```julia
sample2 = rand(Normal(0.2,1.2),30)
```

```julia
histogram(sample1)
histogram!(sample2)
```

```julia
mymodel = compare_dist(sample1,sample2)
```

```julia
chain = Turing.sample(mymodel, NUTS(0.65),1000)
```

```julia
histogram(chain[:deltamu])
```

```julia
function ou(n, sigma, k)
    x = [rand(Normal(0,sigma/sqrt(k)))]
    nd = Normal(0, sigma)
    for i in 1:n-1
        push!(x, (1-k)*x[end] + rand(nd))
    end
    return x
end
```

```julia
xou = ou(100, 1.0, 0.1)
```

```julia
@model function ou_model(oudata)
    k ~ Uniform(0.0001,0.99)
    σ ~ Uniform(0.0001,5)
    oudata[1] ~ Normal(0,σ/sqrt(k))
    for i in 2:length(oudata)
        oudata[i] ~ Normal((1-k)*oudata[i-1],σ)
    end
end
```

```julia
mymodel = ou_model(xou)
```

```julia
chain = Turing.sample(mymodel, NUTS(0.65), 1000)
```

```julia
plot(chain[:k])
```

```julia
plot(chain[:σ])
```
